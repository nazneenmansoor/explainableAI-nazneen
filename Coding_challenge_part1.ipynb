# Example 1: Query a table with SQL and magic commands

```python
# Running this code will query a table in BigQuery and download
# the results to a Pandas DataFrame named `results`.

#Below query selects the data between 2005 and 2009 and station ranging from 725300 and 726300
%%bigquery results --project lively-armor-427812-u5
SELECT * FROM `bigquery-public-data.samples.gsod`
where year between 2005 and 2009 and
station_number between 725300 and 726300
```

```python
# You can view the resulting Pandas DataFrame and work with using the Pandas library.
# https://pandas.pydata.org/docs/getting_started/index.html#getting-started
results
```

```python
#Results into csv file

results.to_csv('snow_data.csv', index=False)
```

```python
from google.cloud import storage

# Set your bucket name
bucket_name = 'coding-challenge-bucket-1'

# Create a storage client
storage_client = storage.Client()

# Upload the file to Google Cloud Storage
bucket = storage_client.bucket(bucket_name)
blob = bucket.blob('snow_data.csv')
blob.upload_from_filename('snow_data.csv')

print(f"File data.csv uploaded to {bucket_name}.")
```
